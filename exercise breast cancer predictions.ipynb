```python
from sklearn.model_selection import GridSearchCV
```

```python
# Create the KNN model
knn = KNeighborsClassifier()

# Set the hyperparameters to be tuned
knn_param_grid = {'n_neighbors': [3, 5, 7, 9]}

# Create the grid search object and fit the data
knn_grid = GridSearchCV(knn, knn_param_grid, cv=10)
knn_grid.fit(X, y)

# Print the best hyperparameters and accuracy score
print("Best KNN Parameters: ", knn_grid.best_params_)
print("Best KNN Accuracy: %0.2f" % (knn_grid.best_score_))

# Create the DT model
dt = DecisionTreeClassifier()

# Set the hyperparameters to be tuned
dt_param_grid = {'max_depth': [3, 5, 7, 9]}

# Create the grid search object and fit the data
dt_grid = GridSearchCV(dt, dt_param_grid, cv=10)
dt_grid.fit(X, y)

# Print the best hyperparameters and accuracy score
print("Best DT Parameters: ", dt_grid.best_params_)
print("Best DT Accuracy: %0.2f" % (dt_grid.best_score_))

# Create the RF model
rf = RandomForestClassifier()

# Set the hyperparameters to be tuned
rf_param_grid = {'n_estimators': [10, 50, 100, 200],
                 'max_depth': [3, 5, 7, 9]}

# Create the grid search object and fit the data
rf_grid = GridSearchCV(rf, rf_param_grid, cv=10)
rf_grid.fit(X, y)

# Print the best hyperparameters and accuracy score
print("Best RF Parameters: ", rf_grid.best_params_)
print("Best RF Accuracy: %0.2f" % (rf_grid.best_score_))
```

The goal of hyperparameter tuning is to find the optimal set of hyperparameters that will maximize the performance of the machine learning model on the task at hand. This is typically done by trying out different combinations of hyperparameters and evaluating their performance on a validation set.
